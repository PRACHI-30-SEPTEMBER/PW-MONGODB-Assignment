## Q1. What is MongoDB? Explain non‑relational databases in short. In which scenarios is MongoDB preferred over SQL databases?

**Answer:**
MongoDB is a NoSQL database that stores data in the form of documents (JSON-like format). Instead of tables and rows, MongoDB uses collections and documents.

A non‑relational database (NoSQL database) stores data in a flexible format and does not require a fixed schema like relational databases.

MongoDB is preferred over SQL databases when:
- Data structure changes frequently.
- Applications require high scalability.
- Large amounts of unstructured or semi‑structured data are used.
- Real‑time analytics or big data applications are needed.

## Q2. State and explain the features of MongoDB.

**Answer:**
Main features of MongoDB:
1. **Document Oriented Storage** – Data is stored as JSON-like documents.
2. **High Performance** – Faster read and write operations.
3. **Scalability** – Supports horizontal scaling using sharding.
4. **Flexible Schema** – Fields can vary from document to document.
5. **Indexing Support** – Improves query performance.
6. **Replication** – Data copies are stored on multiple servers for reliability.

## Q3. Write a code to connect MongoDB to Python. Also create a database and a collection in MongoDB.

```python
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create database
db = client["pwskills_db"]

# Create collection
collection = db["students"]

print("Database and Collection created successfully")
```

## Q4. Insert one record and many records. Use find() and find_one() methods.

```python
# Insert one record
data1 = {"name": "Rahul", "age": 22, "course": "Data Science"}
collection.insert_one(data1)

# Insert multiple records
data_many = [
    {"name": "Priya", "age": 23, "course": "AI"},
    {"name": "Aman", "age": 24, "course": "ML"}
]

collection.insert_many(data_many)

# find_one()
print(collection.find_one())

# find()
for i in collection.find():
    print(i)
```

## Q5. Explain how you can use the find() method to query the MongoDB database.

```python
# Query example
query = {"course": "AI"}

for record in collection.find(query):
    print(record)
```

## Q6. Explain the sort() method with example.

```python
# Sort records by age (ascending)
for record in collection.find().sort("age", 1):
    print(record)
```

## Q7. Explain delete_one(), delete_many(), and drop().

**Answer:**
- **delete_one()** – Deletes a single document matching the condition.
- **delete_many()** – Deletes multiple documents matching the condition.
- **drop()** – Deletes the entire collection.

```python
# delete_one example
collection.delete_one({"name": "Rahul"})

# delete_many example
collection.delete_many({"course": "AI"})

# drop collection
# collection.drop()
```
